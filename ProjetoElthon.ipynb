```python
#CONVERT PDF TO TXT

import PyPDF2
pdffileobj = open('2022.1.pdf', 'rb')
pdfreader = PyPDF2.PdfFileReader(pdffileobj)
x = pdfreader.numPages
for c in range(x):  
  pageobj = pdfreader.getPage(c)
  text = pageobj.extractText()

  file1 = open("2022.2.txt", "a")
  file1.writelines(text)
  file1.close()
```

```python
# NEXT PASS
import json

with open('2022.2.txt') as documento:
  content = documento.readlines()

dice = {}
dice['Cursos'] = {}


for line in range(len(content)):
  # SHOW THE CURSES
  if 'Semestre' in content[line]:
    curse = content[line].strip().split('-')[1]    
    dice['Cursos'][curse] = {} 
    dice['Cursos'][curse]['Professores'] = {}
  

  # SHOW THE TEACHERS
  if 'Vagas Oferecidas' in content[line]:
    teacher = None
    teacher = content[line - 1].strip()
    
    if 'Página' in teacher:
      teacher = teacher.split('Página')[1]        

    if teacher != None:
      try:
        dice['Cursos'][curse]['Professores'][teacher]['Disciplinas']
      except:
        dice['Cursos'][curse]['Professores'][teacher] = {}
        dice['Cursos'][curse]['Professores'][teacher]['Disciplinas'] = []
        
  # SHOW SUBJECTS
  subjects = None
  if 'Vagas Oferecidas' in content[line]:
    subjects = content[line].strip().split('-')[1]

    
  # SHOW HOURS
  hours = None
  if 'Vagas Oferecidas' in content[line]:
    hours = content[line +1].split(':')[1].split('horas')[0].strip()
    
  if subjects != None and hours != None:
    dice['Cursos'][curse]['Professores'][teacher]['Disciplinas'].append((subjects, hours)) 


# FOR JSON
DicesJson = json.dumps(dice)
print(DicesJson)
```
